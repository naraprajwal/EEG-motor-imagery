```python
import os
import sys
import numpy as np
import mne

PROJECT_ROOT = "/Users/prajwalnara/Documents/EEG-motor-imagery"

if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

from src.feature_extraction import extract_bandpower_features

print("Imports successful")
```

```python
EPOCH_ROOT = "/Users/prajwalnara/Documents/EEG-motor-imagery/data/processed/epochs"

FEATURE_ROOT = "/Users/prajwalnara/Documents/EEG-motor-imagery/data/features/band_power_109sub_full64"

os.makedirs(FEATURE_ROOT, exist_ok=True)
```

```python
LABEL_MAP = {
    "rest": 0,
    "left": 1,
    "right": 2,
}
```

```python
total_subjects = 0
successful_subjects = 0
failed_subjects = []
```

```python
for subject in sorted(os.listdir(EPOCH_ROOT)):

    if not subject.startswith("S"):
        continue

    subject_dir = os.path.join(
        EPOCH_ROOT,
        subject
    )

    if not os.path.isdir(subject_dir):
        continue

    print(f"\nProcessing {subject}")

    X_all = []
    y_all = []

    try:

        for file in sorted(os.listdir(subject_dir)):

            if not file.endswith(".fif"):
                continue

            epoch_path = os.path.join(
                subject_dir,
                file
            )

            epochs = mne.read_epochs(
                epoch_path,
                preload=True,
                verbose=False
            )

            X = extract_bandpower_features(
                epochs
            )

            labels = epochs.events[:, -1]

            inverse_event_map = {
                v: k
                for k, v in epochs.event_id.items()
            }

            y = np.array([
                LABEL_MAP[
                    inverse_event_map[label]
                ]
                for label in labels
            ])

            X_all.append(X)
            y_all.append(y)

        X_all = np.vstack(X_all)
        y_all = np.concatenate(y_all)

        save_dir = os.path.join(
            FEATURE_ROOT,
            subject
        )

        os.makedirs(
            save_dir,
            exist_ok=True
        )

        np.save(
            os.path.join(save_dir, "X.npy"),
            X_all
        )

        np.save(
            os.path.join(save_dir, "y.npy"),
            y_all
        )

        print(
            f"Saved {subject} "
            f"X={X_all.shape} "
            f"y={y_all.shape}"
        )

        successful_subjects += 1

    except Exception as e:

        failed_subjects.append(
            (subject, str(e))
        )

        print(
            f"FAILED {subject}: {e}"
        )

    total_subjects += 1
```

```python
print("\n")
print("=" * 60)

print(
    f"Processed : {total_subjects}"
)

print(
    f"Success   : {successful_subjects}"
)

print(
    f"Failed    : {len(failed_subjects)}"
)

print("\nFailed Subjects:")

for item in failed_subjects:
    print(item)
```
