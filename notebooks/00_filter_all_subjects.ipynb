```python
import os
import mne

# ==========================================================
# PATHS
# ==========================================================

RAW_ROOT = "/Users/prajwalnara/Documents/EEG-motor-imagery/data/raw/physionet.org/files/eegmmidb/1.0.0"

FILTERED_ROOT = "/Users/prajwalnara/Documents/EEG-motor-imagery/data/processed/filtered"

os.makedirs(FILTERED_ROOT, exist_ok=True)

# ==========================================================
# FILTER SETTINGS
# ==========================================================

LOW_CUT = 4.0
HIGH_CUT = 40.0
NOTCH_FREQ = 50.0

# ==========================================================
# COUNTERS
# ==========================================================

total_files = 0
success_files = 0
failed_files = []

# ==========================================================
# PROCESS ALL SUBJECTS
# ==========================================================

for subject in sorted(os.listdir(RAW_ROOT)):

    subject_path = os.path.join(RAW_ROOT, subject)

    if not os.path.isdir(subject_path):
        continue

    print("\n" + "=" * 60)
    print(f"Processing {subject}")
    print("=" * 60)

    output_subject_dir = os.path.join(
        FILTERED_ROOT,
        subject
    )

    os.makedirs(output_subject_dir, exist_ok=True)

    edf_files = sorted(
        [
            f
            for f in os.listdir(subject_path)
            if f.lower().endswith(".edf")
        ]
    )

    for edf_file in edf_files:

        total_files += 1

        edf_path = os.path.join(
            subject_path,
            edf_file
        )

        save_name = edf_file.replace(
            ".edf",
            "_filtered_raw.fif"
        )

        save_path = os.path.join(
            output_subject_dir,
            save_name
        )

        # ------------------------------------------
        # Skip already processed files
        # ------------------------------------------

        if os.path.exists(save_path):
            print(f"✓ Already exists: {edf_file}")
            continue

        print(f"Filtering: {edf_file}")

        try:

            # --------------------------------------
            # Try normal EDF loading first
            # --------------------------------------

            try:
                raw = mne.io.read_raw_edf(
                    edf_path,
                    preload=True,
                    verbose=False
                )

            except Exception:

                print("   UTF-8 issue detected -> trying latin1")

                raw = mne.io.read_raw_edf(
                    edf_path,
                    preload=True,
                    encoding="latin1",
                    verbose=False
                )

            # --------------------------------------
            # Channel cleanup
            # --------------------------------------

            raw.rename_channels(
                lambda ch: ch.replace(".", "")
            )

            # --------------------------------------
            # Filtering
            # --------------------------------------

            raw.notch_filter(
                NOTCH_FREQ,
                verbose=False
            )

            raw.filter(
                LOW_CUT,
                HIGH_CUT,
                verbose=False
            )

            # --------------------------------------
            # Save
            # --------------------------------------

            raw.save(
                save_path,
                overwrite=True,
                verbose=False
            )

            success_files += 1

            print(f"   Saved -> {save_name}")

        except Exception as e:

            print(f"   FAILED: {edf_file}")
            print(f"   ERROR : {e}")

            failed_files.append(
                {
                    "subject": subject,
                    "file": edf_file,
                    "error": str(e)
                }
            )

# ==========================================================
# SUMMARY
# ==========================================================

print("\n")
print("=" * 60)
print("FILTERING COMPLETE")
print("=" * 60)

print(f"Total files found     : {total_files}")
print(f"Successfully filtered : {success_files}")
print(f"Failed                : {len(failed_files)}")

if len(failed_files) > 0:

    print("\nFAILED FILES:\n")

    for item in failed_files:

        print(
            f"{item['subject']} | "
            f"{item['file']} | "
            f"{item['error']}"
        )
```
